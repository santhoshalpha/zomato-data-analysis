```python
rest_location = pd.DataFrame()
rest_location['Name'] =   df['location'].unique()
print(rest_location)
```

```python
#to extract the lat and long
#nominator can do that. is a class, is a tool to open street map data to give lat and lon
!pip install geopy
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="app", timeout=None)
```

```python
lat = []
lon = []

for name in rest_location['Name']:
    location = geolocator.geocode(name)    #this extract lat and lon
                                #this is a normal string. we have to convert it into geographical data
        
    if location is None:
        lat.append(np.nan) #way to denote missing value
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
        
        
    
```

```python
rest_location['lat'] = lat
rest_location['lon'] = lon
```
